```python
#crea el pandas
ruta_de_acceso = '/content/drive/MyDrive/Programacion/BigData/Dataset_ventas_AbrilT.csv'

df_ventas = pd.read_csv(ruta_de_acceso)

df_ventas.head(10)
```

```python
#Limpio los Nan que son errores
df_ventas = df_ventas.dropna()
# Cambia los Nan por 0
df_ventas = df_ventas.fillna(0)

df_ventas
```

```python
# Transformo todos los valores no numericos en Nan en las columnas Quantity Ordered y Price Each
df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')

# Elimino todos los Nan y me quedo solo con los valores numericos
df_ventas = df_ventas.dropna()

#reagrupo por producto y ordeno por mas vendido

# Realizo un agrupamiento de los datos donde a la cantidad ordenada le aplico la suma, a el precio el promedio y lo agrupo con los productos.
productos_vendidos_ordenados = df_ventas.groupby('Producto').agg({
    'Cantidad Pedida': 'sum',
    'Precio Unitario': 'mean'
})
 #Aca se puede hacer muchas operaciones .sum() .min() .max() .std() etc

# Ordeno los de mayor a menor segun la cantidad de productos vendidos
#productos_vendidos_ordenados = productos_vendidos.sort_values(ascending=False)

productos_vendidos_ordenados
```

```python
#extraigo los datos del df y los paso a arrays
y_values = productos_vendidos_ordenados['Cantidad Pedida'].tolist()
x_values = df_ventas['Producto'].unique()
print (y_values)
print (x_values)
```

```python
#Creo el objeto figura
fig = go.Figure()

#Agrego grafico en mi objeto figura
fig.add_trace(go.Bar(
        x = x_values,
        y= y_values,
        name = 'Unidades Vendidas',
        marker = dict(color= 'Light Blue')
)
)

# Configuro titulos y ejes del grafico
fig.update_layout(
    title = 'Unidades Vendidas',
    xaxis = dict(title='Productos'),
    yaxis = dict(title='Unidades')

)
```
