```python
#crea el pandas
ruta_de_acceso = '/content/drive/MyDrive/Programacion/BigData/Dataset_ventas_AbrilT.csv'

df_ventas = pd.read_csv(ruta_de_acceso)

df_ventas.head(10)
```

```python
#Limpio los Nan que son errores
df_ventas = df_ventas.dropna()
# Cambia los Nan por 0
df_ventas = df_ventas.fillna(0)

df_ventas
```

```python
# Transformo todos los valores no numericos en Nan en las columnas Quantity Ordered y Price Each
df_ventas['Cantidad Pedida'] = pd.to_numeric(df_ventas['Cantidad Pedida'], errors='coerce')  #El errors= 'coerce' transforma los valores no numericos a NaN'
df_ventas['Precio Unitario'] = pd.to_numeric(df_ventas['Precio Unitario'], errors='coerce')

# Elimino todos los Nan y me quedo solo con los valores numericos
df_ventas = df_ventas.dropna()

#reagrupo por producto y ordeno por mas vendido

# Realizo un agrupamiento de los datos donde a la cantidad ordenada le aplico la suma, a el precio el promedio y lo agrupo con los productos.
productos_vendidos_ordenados = df_ventas.groupby('Producto').agg({
    'Cantidad Pedida': 'sum',
    'Precio Unitario': 'mean'
})
 #Aca se puede hacer muchas operaciones .sum() .min() .max() .std() etc

# Ordeno los de mayor a menor segun la cantidad de productos vendidos
#productos_vendidos_ordenados = productos_vendidos.sort_values(ascending=False)

productos_vendidos_ordenados
```

```python
#extraigo los datos del df y los paso a arrays
y_values = productos_vendidos_ordenados['Cantidad Pedida'].tolist()
x_values = df_ventas['Producto'].unique()
print (y_values)
print (x_values)
```

```python
#Creo el objeto figura
fig = go.Figure()

#Agrego grafico en mi objeto figura
fig.add_trace(go.Bar(
        x = x_values,
        y= y_values,
        name = 'Unidades Vendidas',
        marker = dict(color= 'Light Blue')
)
)

# Configuro titulos y ejes del grafico
fig.update_layout(
    title = 'Unidades Vendidas',
    xaxis = dict(title='Productos'),
    yaxis = dict(title='Unidades')

)
```

# **Graficar que porcentaje representa cada producto en mi total de ventas**

```python
#convierto los valores a porcentajes
# calculamos la columna porcentaje, la cual se añade automaticmente # al dataframe accesorios. Dividimos la cantidad entre la suma de
# la columna y la multiplicamos por 100

productos_vendidos_ordenados['porcentaje'] = (productos_vendidos_ordenados['Cantidad Pedida']/ productos_vendidos_ordenados['Cantidad Pedida'].sum())*100

productos_vendidos_ordenados
```

```python
#extraigo valores
y_values = productos_vendidos_ordenados['porcentaje'].tolist()
x_values = df_ventas['Producto'].unique()

fig = px.pie(values = y_values,
             names = x_values,
             title = 'Productos vendidos')

fig.show()
```

# **Analisis de Ventas en el tiempo**

```python
# agrego una nueva columna "fecha de compra" en formato fecha de python
df_ventas['Fecha de compra'] = pd.to_datetime(df_ventas['Fecha de Pedido'], dayfirst=True)

df_ventas

```

```python
# Realizamos una funcion para extraer las caracteristicas del datetime, para esto crea 2 columnas nuevas Hora y Fecha
def extraer_datos(datos):
  datos['Hora'] =  datos['Fecha de compra'].dt.hour
  datos['Fecha'] = datos['Fecha de compra'].dt.date

  return datos

df_ventas = extraer_datos(df_ventas)

df_ventas
```

```python
#separo los datos de Dia en un array para graficar
#ventas_por_dia = df_ventas.groupby('Fecha')['Cantidad Pedida'].sum()

ventas_por_dia = df_ventas.groupby('Fecha').agg({
    'Cantidad Pedida': 'sum',
})

fechas= list(ventas_por_dia.index)
cantidad = ventas_por_dia['Cantidad Pedida'].tolist()
print (cantidad)
```

```python


fig = px.line(x = fechas, y = cantidad,
              title = "Título del gráfico de líneas",
              labels = {'x': 'Tiempo', 'y':'Ventas'},
              markers = True)


fig.show()
```

```python
fig = px.line(x = fechas, y = cantidad,
              text= cantidad,
              title = "Título del gráfico de líneas",
              labels = {'x': 'Tiempo', 'y':'Ventas'},
              markers = True)


fig.show()
```
